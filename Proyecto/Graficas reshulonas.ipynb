```python
import csv
import operator
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from matplotlib.patches import Patch

lista =[]
 
with open('aristas.csv', newline='') as File:  
    next(File)
    reader = csv.reader(File)
    for row in reader:
        lista.append([row[0], row[1], int(row[6])])
lista.sort(key=lambda x: x[2], reverse = True)
```

```python
aristas = {}
for e in lista[:12]:
    key = e[0] + " - " + e[1]
    aristas[key] = e[2]
```

```python
# plt styling parameters
plt.style.use('seaborn')
plt.rcParams['figure.figsize'] = (8.0, 6.0)
plt.rcParams["patch.force_edgecolor"] = True

# create a color palette the length of the dataframe
colors = sns.color_palette('husl', n_colors=len(aristas))
index = []
for i in range(0, 12): 
    index.append(i)

# plot
plt.bar(index ,aristas.values(), color=colors)

#sns.barplot(x= aristas.keys(), y='values', data=aristas.values(), palette=colors)

# create color map with colors and df.names
cmap = dict(zip(aristas.keys(), colors))

# create the rectangles for the legend
patches = [Patch(color=v, label=k) for k, v in cmap.items()]

# place the legend to the right of the plot
plt.legend(handles=patches, bbox_to_anchor=(1.04, 0.5), loc='center left', borderaxespad=0)
```
